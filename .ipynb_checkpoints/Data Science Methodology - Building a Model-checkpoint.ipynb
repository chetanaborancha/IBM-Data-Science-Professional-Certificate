```python
import pandas as pd # import library to read data into dataframe
pd.set_option("display.max_columns", None)
import numpy as np # import numpy library
import re # import library for regular expression
import random # library for random number generation
```

```python
recipes = pd.read_csv("https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DS0103EN/labs/data/recipes.csv")

print("Data read into dataframe!") # takes about 30 seconds
```

```python
#Preprocessing

#fix name of the column displaying the cuisine
column_names = recipes.columns.values
column_names[0]="cuisines"
recipes.columns=column_names

#convert cuisine names to lower case
recipes["cuisines"] = recipes["cuisines"].str.lower()
```

```python
# make the cuisine names consistent


recipes.loc[recipes["cuisines"] == "austria", "cuisines"] = "austrian"
recipes.loc[recipes["cuisines"] == "belgium", "cuisines"] = "belgian"
recipes.loc[recipes["cuisines"] == "china", "cuisines"] = "chinese"
recipes.loc[recipes["cuisines"] == "canada", "cuisines"] = "canadian"
recipes.loc[recipes["cuisines"] == "netherlands", "cuisines"] = "dutch"
recipes.loc[recipes["cuisines"] == "france", "cuisines"] = "french"
recipes.loc[recipes["cuisines"] == "germany", "cuisines"] = "german"
recipes.loc[recipes["cuisines"] == "india", "cuisines"] = "indian"
recipes.loc[recipes["cuisines"] == "indonesia", "cuisines"] = "indonesian"
recipes.loc[recipes["cuisines"] == "iran", "cuisines"] = "iranian"
recipes.loc[recipes["cuisines"] == "italy", "cuisines"] = "italian"
recipes.loc[recipes["cuisines"] == "japan", "cuisines"] = "japanese"
recipes.loc[recipes["cuisines"] == "israel", "cuisines"] = "jewish"
recipes.loc[recipes["cuisines"] == "korea", "cuisines"] = "korean"
recipes.loc[recipes["cuisines"] == "lebanon", "cuisines"] = "lebanese"
recipes.loc[recipes["cuisines"] == "malaysia", "cuisines"] = "malaysian"
recipes.loc[recipes["cuisines"] == "mexico", "cuisines"] = "mexican"
recipes.loc[recipes["cuisines"] == "pakistan", "cuisines"] = "pakistani"
recipes.loc[recipes["cuisines"] == "philippines", "cuisines"] = "philippine"
recipes.loc[recipes["cuisines"] == "scandinavia", "cuisines"] = "scandinavian"
recipes.loc[recipes["cuisines"] == "spain", "cuisines"] = "spanish_portuguese"
recipes.loc[recipes["cuisines"] == "portugal", "cuisines"] = "spanish_portuguese"
recipes.loc[recipes["cuisines"] == "switzerland", "cuisines"] = "swiss"
recipes.loc[recipes["cuisines"] == "thailand", "cuisines"] = "thai"
recipes.loc[recipes["cuisines"] == "turkey", "cuisines"] = "turkish"
recipes.loc[recipes["cuisines"] == "vietnam", "cuisines"] = "vietnamese"
recipes.loc[recipes["cuisines"] == "uk-and-ireland", "cuisines"] = "uk-and-irish"
recipes.loc[recipes["cuisines"] == "irish", "cuisines"] = "uk-and-irish"
```

```python
recipes_counts = recipes["cuisines"].value_counts()
cuisines_indices = recipes_counts > 50
```

```python
cuisines_to_keep = list(np.array(recipes_counts.index.values)[np.array(cuisines_indices)])
recipes = recipes.loc[recipes["cuisines"].isin(cuisines_to_keep)]
```

```python
recipes.head()
```

```python
recipes = recipes.replace(to_replace="Yes",value=1)
recipes = recipes.replace(to_replace="No",value=0)
```

```python
recipes.head()
```

```python
#decision tree using scikit learn

%matplotlib inline
from sklearn import tree
from sklearn.metrics import accuracy_score, confusion_matrix

import matplotlib.pyplot as plt
!conda install python-graphviz --yes
import graphviz


from sklearn.tree import export_graphviz

import itertools
```

Here, we are creating a decision tree for the recipes for just some of the Asian (Korean, Japanese, Chinese, Thai) and Indian cuisines. The reason for this is because the decision tree does not run well when the data is biased towards one cuisine, in this case American cuisines. One option is to exclude the American cuisines from our analysis or just build decision trees for different subsets of the data. Let's go with the latter solution.

Let's build our decision tree using the data pertaining to the Asian and Indian cuisines and name our decision tree - tree.

```python
# select subset of cuisines
asian_indian_recipes = recipes[recipes.cuisines.isin(["korean", "japanese", "chinese", "thai", "indian"])]
cuisines = asian_indian_recipes["cuisines"]
ingredients = asian_indian_recipes.iloc[:,1:]
```

```python
dtree = tree.DecisionTreeClassifier(max_depth=3)
dtree.fit(ingredients, cuisines)

print("Decision tree model saved to tree!")
```

```python
export_graphviz(dtree,
                feature_names=list(ingredients.columns.values),
                out_file="dtree.dot",
                class_names=np.unique(cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("dtree.dot") as dtree_image:
    dtree_graph = dtree_image.read()
graphviz.Source(dtree_graph)
```

The decision tree learned:

- If a recipe contains cumin and fish and no yoghurt, then it is most likely a Thai recipe
- If a recipe contains cumin but no fish and no soy_sauce, then it is most likely an Indian recipe

To evaluate our model of Asian and Indian cuisines, we will split our dataset into a training set and a test set. We will build the decision tree using the training set. Then, we will test the model on the test set and compare the cuisines that the model predicts to the actual cuisines.

Let's first create a new dataframe using only the data pertaining to the Asian and the Indian cuisines, and let's call the new dataframe bamboo.

```python
bamboo = recipes[recipes.cuisines.isin(["korean", "japanese", "chinese", "thai", "indian"])]
```

```python
bamboo["cuisines"].value_counts()
```

```python
#Let's remove 30 recipes from each cuisine to use as the test set, and let's name this test set bamboo_test
sample_n = 30
```

```python
#Create a dataframe containing 30 recipes from each cuisine, selected randomly.
random.seed(1234) # set random seed
bamboo_test = bamboo.groupby("cuisines", group_keys=False).apply(lambda x: x.sample(sample_n))

bamboo_test_ingredients = bamboo_test.iloc[:,1:] # ingredients
bamboo_test_cuisines = bamboo_test["cuisines"] # corresponding cuisines or labels
```

```python
bamboo_test
```

```python
#Check that there are 30 recipes for each cuisine.

bamboo_test["cuisines"].value_counts()
```

```python
#Next, let's create the training set by removing the test set from the bamboo dataset, and let's call the training set bamboo_train.

bamboo_test_index = bamboo.index.isin(bamboo_test.index)
bamboo_train = bamboo[~bamboo_test_index]

bamboo_train_ingredients = bamboo_train.iloc[:,1:] # ingredients
bamboo_train_cuisines = bamboo_train["cuisines"] # corresponding cuisines or labels
```

```python
bamboo_test_index
```

```python
bamboo_train["cuisines"].value_counts()
```

```python
#Let's build the decision tree using the training set, bamboo_train, and name the generated tree bamboo_train_tree for prediction

bamboo_train_tree = tree.DecisionTreeClassifier(max_depth=15)
bamboo_train_tree.fit(bamboo_train_ingredients, bamboo_train_cuisines)

print("Decision tree model saved to bamboo_train_tree!")
```

```python
export_graphviz(bamboo_train_tree,
                feature_names=list(bamboo_train_ingredients.columns.values),
                out_file="bamboo_train_tree.dot",
                class_names=np.unique(bamboo_train_cuisines),
                filled=True,
                node_ids=True,
                special_characters=True,
                impurity=False,
                label="all",
                leaves_parallel=False)

with open("bamboo_train_tree.dot") as bamboo_train_tree_image:
    bamboo_train_tree_graph = bamboo_train_tree_image.read()
graphviz.Source(bamboo_train_tree_graph)
```

```python
bamboo_pred_cuisines = bamboo_train_tree.predict(bamboo_test_ingredients)
```

To quantify how well the decision tree is able to determine the cuisine of each recipe correctly, we will create a confusion matrix which presents a nice summary on how many recipes from each cuisine are correctly classified. It also sheds some light on what cuisines are being confused with what other cuisines.

So let's go ahead and create the confusion matrix for how well the decision tree is able to correctly classify the recipes in bamboo_test.

```python
test_cuisines = np.unique(bamboo_test_cuisines)
bamboo_confusion_matrix = confusion_matrix(bamboo_test_cuisines, bamboo_pred_cuisines, test_cuisines)
title = 'Bamboo Confusion Matrix'
cmap = plt.cm.Blues

plt.figure(figsize=(8,6))
bamboo_confusion_matrix = (
    bamboo_confusion_matrix.astype('float') / bamboo_confusion_matrix.sum(axis=1)[:, np.newaxis]
    ) * 100

plt.imshow(bamboo_confusion_matrix, interpolation='nearest', cmap=cmap)
plt.title(title)
plt.colorbar()
tick_marks = np.arange(len(test_cuisines))
plt.xticks(tick_marks, test_cuisines)
plt.yticks(tick_marks, test_cuisines)

fmt = '.2f'
thresh = bamboo_confusion_matrix.max() / 2.
for i, j in itertools.product(range(bamboo_confusion_matrix.shape[0]), range(bamboo_confusion_matrix.shape[1])):
    plt.text(j, i, format(bamboo_confusion_matrix[i, j], fmt),
             horizontalalignment="center",
             color="white" if bamboo_confusion_matrix[i, j] > thresh else "black")


plt.ylabel('True label')
plt.xlabel('Predicted label')
plt.ylim(-0.5, 4.5)

plt.show()
plt.tight_layout()
```

The rows represent the actual cuisines from the dataset and the columns represent the predicted ones. Each row should sum to 100%. According to this confusion matrix, we make the following observations:

Using the last row in the confusion matrix, 60% of the Chinese recipes in bamboo_test were correctly classified by our decision tree whereas 23.33% of the Chinese recipes were misclassified as Korean and 3.33% were misclassified as Thai.

Using the Indian row, 96.67% of the Indian recipes in bamboo_test were correctly classified by our decision tree and 3.33% of the Indian recipes were misclassified as Korean.
